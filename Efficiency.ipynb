```python
import ROOT
```

```python
f=ROOT.TFile("out.root")
flux=f.Get("flux")
generated=f.Get("generated")
```

```python
denom= ROOT.TH2D("denom", "denom",100, -.4, .4, 100, -.4,.4)
num= ROOT.TH2D("num", "Efficiency; px/pz; py/pz",100, -.4, .4, 100, -.4,.4)

for fl, gen in zip(flux, generated):
    ux = gen.px[0]/gen.pz[0]
    uy = gen.py[0]/gen.pz[0]
    if(len(fl.id)>0):
        num.Fill(ux, uy)
    denom.Fill(ux, uy)
num.Divide(denom)
```

```python
c1=ROOT.TCanvas("c1", "c1", 800, 800)
num.Draw("COLZ")
ROOT.gStyle.SetOptStat(0)
c1.Draw()
```
